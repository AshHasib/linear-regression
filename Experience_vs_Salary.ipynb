# Experience vs Salary 
This is a linear regression example on a sample dataset from Kaggle. Here, we have to predict the expected salary of an employee based on his years of experience. The dataset contains two columns of data where each column represents Experience and Salary respectively.

## Importing the necessary libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

## Reading the dataset from CSV file

```python
dataframe = pd.read_csv('Salary_Data.csv')
X = dataframe.iloc[:, :-1].values
Y = dataframe.iloc[:,1].values

#print(X)
#print(Y)
```

## A random Test-train split between the data in the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 1/3,random_state=0)
```

## Creating the model
Here we have declared our very own simple Linear Regression model and fit the training data found in the previous step into the model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

## Predicting values using test data

```python
predictions = regressor.predict(X_test)
print(Y_test)
print(predictions)
```

## Visualizing the training best-fit line

```python
plt.scatter(X_train, Y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary v Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualizing the best-fit line in test data

```python
plt.scatter(X_test, Y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary v Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
